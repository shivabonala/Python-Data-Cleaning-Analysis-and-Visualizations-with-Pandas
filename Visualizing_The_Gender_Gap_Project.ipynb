## Storytelling Through Data Visualization with Pandas

In this notebook, we will explore a new dataset by storytelling data visualizations using pandas. Also, we will improve plot aesthitics by adding color, layout and annotations. We'll be working with a dataset containing the percentage of bachelor's degrees granted to women from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category.Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. We will explore how we can communicate the nuanced narrative of gender gap using effective data visualization.


## 1. Import the plotting libraries 

```python
# Import pandas and matplotlib into the environmentfa
# Run the Jupyter magic %matplotlib inline so that plots are displayed inline
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read dataset into pandas
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
# return the first row formatted as a table
women_degrees.iloc[0]
```

```python
#  Setting Line Color Using RGB
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
```

## 2.Compare six degrees

```python
#  Order the charts by decreasing ending gender gap and populate the list
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
```

```python
# use plot layut of single row with multiple columns
# use linewidth parameter to specify the line width. 
# use c parameter to set lne colors
# Romove all the tick marks using axes fucntion tick_params
# Hide all the spines by creating axes instance and using set_visible() method for each spine object dictionary
# set xlimit, ylimit and title
# Add text annotation only to first and last plot using axes.text() method and pass coordinate values



fig = plt.figure(figsize=(18, 3))

for sp in range(6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    if sp==0:
        ax.text(2005,87,"Men")
        ax.text(2002,8,"Women")
    elif sp==5:
        ax.text(2005,62,"Men")
        ax.text(2001,35,"Women")
        
plt.show();
```

## 2. Compare all degrees


```python
#  List of all catagories
all_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science',
            'Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History',
            'Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
#  use subplot grid layout of 6 rows by 3 columns
fig = plt.figure(figsize=(18, 30))

for sp in range(18):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[all_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[all_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(all_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
   
    if all_cats[sp] == 'Psychology':
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif all_cats[sp] == 'Engineering':
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
    elif all_cats[sp] == 'Foreign Languages':
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
    elif all_cats[sp] == 'Health Professions':
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif all_cats[sp] == 'Architecture':
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        
plt.show();
```

## 3. Hiding x-axis labels

```python
# Disable the x-axis labels for all line charts except the bottommost line chart

fig = plt.figure(figsize=(18, 30))

for sp in range(18):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[all_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[all_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(all_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
   
    if all_cats[sp] == 'Psychology':
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif all_cats[sp] == 'Engineering':
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
    elif all_cats[sp] == 'Foreign Languages':
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
    elif all_cats[sp] == 'Health Professions':
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif all_cats[sp] == 'Architecture':
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        
    if all_cats[sp] in ('Agriculture','Business', 'Architecture'):
        ax.tick_params(labelbottom="on")
        
plt.show();
```


## 4. Setting y-axis labels

```python
## set y-axis labes as 0 and 100

fig = plt.figure(figsize=(18, 30))

for sp in range(18):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[all_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[all_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(all_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
   
    if all_cats[sp] == 'Psychology':
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif all_cats[sp] == 'Engineering':
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
    elif all_cats[sp] == 'Foreign Languages':
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
    elif all_cats[sp] == 'Health Professions':
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif all_cats[sp] == 'Architecture':
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        
    if all_cats[sp] in ('Agriculture','Business', 'Architecture'):
        ax.tick_params(labelbottom="on")
        
plt.show();
```

## 5. Adding a horizontal line 

```python
#  Add horizontal line across an entire subplot 
# use axhline() method and set the color and transparency using c and alpha parameter
fig = plt.figure(figsize=(18, 30))
c_hline =(171/255, 171/255, 171/255)

for sp in range(18):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[all_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[all_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(all_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
    ax.axhline(50, c=c_hline, alpha=0.4)
   
    if all_cats[sp] == 'Psychology':
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif all_cats[sp] == 'Engineering':
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
    elif all_cats[sp] == 'Foreign Languages':
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
    elif all_cats[sp] == 'Health Professions':
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif all_cats[sp] == 'Architecture':
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        
    if all_cats[sp] in ('Agriculture','Business', 'Architecture'):
        ax.tick_params(labelbottom="on")
        
plt.show();
```
